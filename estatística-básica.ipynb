```python
import pandas as pd
```

```python
import numpy as np
```

```python
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
```

```python
np.set_printoptions(precision=2, suppress=True)
```

```python
lista_idades = [26,30,32,22,26,35,40,20,43,31,23]
```

```python
soma= np.sum(lista_idades)
print("Soma: ", soma)
```

```python
qt= len(lista_idades)
print("Quantidade de idades: ", qt)
```

```python
media= np.sum(lista_idades)/len(lista_idades)
print("Media: ", media)
```

```python
media= np.mean(lista_idades)
print("Media aritmetica: ", media)
```

```python
mediana= np.median(lista_idades)
print("Mediana: ", mediana)
```

```python
df= pd.read_excel('planilha_modulo3.xlsx')
```

```python
Media= df['IDADE'].mean()
print("Media: ", Media)
```

```python
mediana= df['IDADE'].median()
print("Mediana: ", mediana)
```

```python
moda= df['IDADE'].mode()
print("Moda: ", moda)
```

```python
dp= df['IDADE'].std()
print("Desvio padrão: ", dp)
```

```python
vmin= df['IDADE'].min()
print("Valor mínimo: ", vmin)
```

```python
vmax= df['IDADE'].max()
print("Valor máximo: ", vmax)
```

```python
fem= df[df['GENERO']=='Feminino']['IDADE'].mean()
print("Média da idade das pessoas que responderam ser do gênero feminino: ", fem)
```

```python
masc= df[df['GENERO']=='Masculino']['IDADE'].mean()
print("Média da idade das pessoas que responderam ser do gênero masculino: ", masc)
```

```python
mascsal= df[df['GENERO']=='Masculino']['SALARIO'].mean()
print("Estatística da coluna de salário das pessoas que responderam ser do gênero masculino: ", mascsal)
```
