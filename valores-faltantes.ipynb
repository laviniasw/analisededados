```python
import pandas as pd
```

```python
import numpy as np
```

```python
df= pd.read_excel('planilha_modulo3.xlsx')
```

```python
df.groupby('GENERO', dropna=False)['ID'].nunique()
```

```python
df['GENERO']=df['GENERO'].fillna('Prefiro não informar')
```

```python
df[df['IDADE'].isnull()]['FAIXA IDADE'].value_counts()
```

```python
md1= df[df['FAIXA IDADE']=='17-21']['IDADE'].mean()
print("Média das idades entre 17 e 21: ", md1)
```

```python
df.loc[(df['FAIXA IDADE']=='17-21') & (df['IDADE'].isnull()), 'IDADE']
```

```python
df.loc[(df['FAIXA IDADE']=='17-21') & (df['IDADE'].isnull()), 'IDADE'] = md1
```

```python
media= df['IDADE'].mean()
print("Media geral: ", media)
```

```python
df.loc[(df['FAIXA IDADE']=='55+') & (df['IDADE'].isnull()), 'IDADE'] = media
```

```python
medianasal= df['SALARIO'].median()
print("A Mediana da coluna do salário é: ", medianasal)
```

```python
df.loc[(df['SALARIO'].isnull()), 'SALARIO'] = medianasal
```
